```python
import numpy as np
import matplotlib.pyplot as plt
neuronal_data = np.genfromtxt(open("fig1_2_neuronal_data.csv", "rb"), delimiter=",")

time = np.linspace(0,221,len(neuronal_data[0]))

AFDL = neuronal_data[0]
AFDL = (AFDL - np.mean(AFDL))/np.std(AFDL)

AFDR = neuronal_data[1]
AFDR = (AFDR - np.mean(AFDR))/np.std(AFDR)

AIBL = neuronal_data[2]
AIBL = (AIBL - np.mean(AIBL))/np.std(AIBL)

AVAL = neuronal_data[3]
AVAL = (AVAL - np.mean(AVAL))/np.std(AVAL)

AVBL = neuronal_data[4]
AVBL = (AVBL - np.mean(AVBL))/np.std(AVBL)

AVDL = neuronal_data[5]
AVDL = (AVDL - np.mean(AVDL))/np.std(AVDL)

AVDR = neuronal_data[6]
AVDR = (AVDR - np.mean(AVDR))/np.std(AVDR)

RIML = neuronal_data[7]
RIML = (RIML - np.mean(RIML))/np.std(RIML)

RMEL = neuronal_data[8]
RMEL = (RMEL - np.mean(RMEL))/np.std(RMEL)
```
