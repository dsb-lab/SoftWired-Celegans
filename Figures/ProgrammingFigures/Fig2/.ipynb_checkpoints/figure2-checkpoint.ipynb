```python
import numpy as np
import matplotlib.pyplot as plt

fSize = 18; lPad = 20; lim=8

inhibition = np.genfromtxt(open("fig2_inhibition_iterations.csv", "rb"), delimiter=",")
fitness = np.genfromtxt(open("fig2_fitness_iterations.csv", "rb"), delimiter=",")

fit_AIBL = np.genfromtxt(open("fig2_fit_AIBL.csv", "rb"), delimiter=",")
fit_AVAL = np.genfromtxt(open("fig2_fit_AVAL.csv", "rb"), delimiter=",")
fit_AVBL = np.genfromtxt(open("fig2_fit_AVBL.csv", "rb"), delimiter=",")
fit_AVDL = np.genfromtxt(open("fig2_fit_AVDL.csv", "rb"), delimiter=",")
fit_AVDR = np.genfromtxt(open("fig2_fit_AVDR.csv", "rb"), delimiter=",")
fit_RIML = np.genfromtxt(open("fig2_fit_RIML.csv", "rb"), delimiter=",")
fit_RMEL = np.genfromtxt(open("fig2_fit_RMEL.csv", "rb"), delimiter=",")

fitness_AIBL = np.genfromtxt(open("fig2_fitness_AIBL_iterations.csv", "rb"), delimiter=",")
fitness_AVAL = np.genfromtxt(open("fig2_fitness_AVAL_iterations.csv", "rb"), delimiter=",")
fitness_AVBL = np.genfromtxt(open("fig2_fitness_AVBL_iterations.csv", "rb"), delimiter=",")
fitness_AVDL = np.genfromtxt(open("fig2_fitness_AVDL_iterations.csv", "rb"), delimiter=",")
fitness_AVDR = np.genfromtxt(open("fig2_fitness_AVDR_iterations.csv", "rb"), delimiter=",")
fitness_RIML = np.genfromtxt(open("fig2_fitness_RIML_iterations.csv", "rb"), delimiter=",")
fitness_RMEL = np.genfromtxt(open("fig2_fitness_RMEL_iterations.csv", "rb"), delimiter=",")

time = np.linspace(0,221,len(fit_AIBL))
iterations = range(1,101)


fit_AIBL = (fit_AIBL - np.mean(fit_AIBL))/np.std(fit_AIBL)
fit_AVAL = (fit_AVAL - np.mean(fit_AVAL))/np.std(fit_AVAL)
fit_AVBL = (fit_AVBL - np.mean(fit_AVBL))/np.std(fit_AVBL)
fit_AVDL = (fit_AVDL - np.mean(fit_AVDL))/np.std(fit_AVDL)
fit_AVDR = (fit_AVDR - np.mean(fit_AVDR))/np.std(fit_AVDR)
fit_RIML = (fit_RIML - np.mean(fit_RIML))/np.std(fit_RIML)
fit_RMEL = (fit_RMEL - np.mean(fit_RMEL))/np.std(fit_RMEL)
```

```python
width = 9#6.625
height = 23
nplots = 22
fig3 = plt.figure(figsize=(width,height),constrained_layout=False)
alpha = 0.5
labelPos = -0.075
color = "black"


gs = fig3.add_gridspec(nplots, 1)

###
### First plot
###

f3_ax1 = fig3.add_subplot(gs[1])
f3_ax1.plot(time, fit_AIBL, color = "red",alpha = 0.6)
f3_ax1.set_ylim(-lim,lim)
f3_ax1.set_xlim(0,221)
f3_ax1.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax1.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax1.grid(alpha = alpha)
f3_ax1.set_title('signal comparison',fontsize = fSize+2)
f3_ax1_2 = f3_ax1.twinx()
f3_ax1_2.set_ylabel('AIBL',fontsize = fSize, rotation=270,labelpad=lPad)
f3_ax1_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax2 = fig3.add_subplot(gs[2])
f3_ax2.plot(time, fit_AVAL, color = "red",alpha = 0.6)
f3_ax2.set_ylim(-lim,lim)
f3_ax2.set_xlim(0,221)
f3_ax2.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax2.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax2.grid(alpha = alpha)
f3_ax2_2 = f3_ax2.twinx()
f3_ax2_2.set_ylabel('AVAL',fontsize = fSize, rotation=270,labelpad=lPad)
f3_ax2_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax3 = fig3.add_subplot(gs[3])
f3_ax3.plot(time, fit_AVBL, color = "red",alpha = 0.6)
f3_ax3.set_ylim(-lim,lim)
f3_ax3.set_xlim(0,221)
f3_ax3.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax3.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax3.grid(alpha = alpha)
f3_ax3_2 = f3_ax3.twinx()
f3_ax3_2.set_ylabel('AVBL',fontsize = fSize, rotation=270,labelpad=lPad)
f3_ax3_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax4 = fig3.add_subplot(gs[4])
f3_ax4.plot(time, fit_AVDL, color = "red",alpha = 0.6)
f3_ax4.set_ylabel('activity (a.u.)',fontsize = fSize)
f3_ax4.get_yaxis().set_label_coords(labelPos,0.5)
f3_ax4.set_ylim(-lim,lim)
f3_ax4.set_xlim(0,221)
f3_ax4.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax4.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax4.grid(alpha = alpha)
f3_ax4_2 = f3_ax4.twinx()
f3_ax4_2.set_ylabel('AVDL',fontsize = fSize, rotation=270,labelpad=lPad)
f3_ax4_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax5 = fig3.add_subplot(gs[5])
f3_ax5.plot(time, fit_AVDR, color = "red",alpha = 0.6)
f3_ax5.set_ylim(-lim,lim)
f3_ax5.set_xlim(0,221)
f3_ax5.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax5.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax5.grid(alpha = alpha)
f3_ax5_2 = f3_ax5.twinx()
f3_ax5_2.set_ylabel('AVDR',fontsize = fSize, rotation=270,labelpad=lPad)
f3_ax5_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax6 = fig3.add_subplot(gs[6])
f3_ax6.plot(time, fit_RIML, color = "red",alpha = 0.6)
f3_ax6.set_ylim(-lim,lim)
f3_ax6.set_xlim(0,221)
f3_ax6.tick_params(axis='x', direction='in', bottom=False, top=False, labelbottom=False) 
f3_ax6.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax6.grid(alpha = alpha)
f3_ax6_2 = f3_ax6.twinx()
f3_ax6_2.set_ylabel('RIML',fontsize = fSize, rotation=270,labelpad=lPad-2)
f3_ax6_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 

f3_ax7 = fig3.add_subplot(gs[7])
f3_ax7.plot(time, fit_RMEL, label='simulated', color = "red",alpha = 0.6)
f3_ax7.set_ylim(-lim,lim)
f3_ax7.set_xlim(0,221)
f3_ax7_2 = f3_ax7.twinx()
f3_ax7_2.set_ylabel('RMEL',fontsize = fSize, rotation=270,labelpad=lPad-2)
f3_ax7_2.tick_params(axis='y', direction='in', bottom=False, top=False, labelright=False) 
f3_ax7.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax7.tick_params(axis='x', direction='in', bottom=True, top=False, labelbottom=True, labelsize=fSize-2)
f3_ax7.set_xlabel('time (s)',fontsize = fSize)
f3_ax7.grid(alpha = alpha)
f3_ax7.legend(loc='center',bbox_to_anchor=(0.5,-1.5),fontsize=fSize,ncol=2)


f3_ax8 = fig3.add_subplot(gs[10:12])
f3_ax8.plot(iterations, fitness, color=color)
f3_ax8.set_title('optimal network evolution',fontsize = fSize+2)
f3_ax8.set_ylim(0.5,0.8)
f3_ax8.set_xlim(1,100)
f3_ax8.grid(alpha = alpha)
f3_ax8.set_ylabel('fitness',fontsize = fSize)
f3_ax8.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax8.tick_params(axis='x', direction='in', bottom=True, top=False, labelbottom=False, labelsize=fSize-2)
f3_ax8.get_yaxis().set_label_coords(labelPos,0.5)


f3_ax16 = fig3.add_subplot(gs[12:14])
#f3_ax16.plot(iterations, inhibition, color = "#ff7f0e")
f3_ax16.plot(iterations, inhibition, color=color)
f3_ax16.set_ylim(0,60)
f3_ax16.set_xlim(1,100)
f3_ax16.grid(alpha = alpha)
f3_ax16.set_xlabel('iterations',fontsize = fSize)
f3_ax16.set_ylabel('% inhibition',fontsize = fSize)
f3_ax16.tick_params(axis='y', direction='in', bottom=False, top=False, labelleft=True, labelsize=fSize-2) 
f3_ax16.tick_params(axis='x', direction='in', bottom=True, top=False, labelbottom=True, labelsize=fSize-2)
f3_ax16.get_yaxis().set_label_coords(labelPos,0.5)
plt.savefig("fig2b.pdf",bbox_inches='tight')
```
